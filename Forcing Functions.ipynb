```python
%load_ext autoreload
%autoreload 2

# Third-Party Imports
import numpy as np
from numpy import sin, cos
import matplotlib.pyplot as plt

# Package Imports
from sindy_bvp.bvp_solver import get_forcings
```

```python
##############
## SET UP x ##
##############

# Define the independent variable bounds and discrete increments
x_min = 0.0; x_max = 10; dx = 0.01;
x = np.linspace(x_min, x_max, int((x_max-x_min)/dx)+1)
```

```python
#######################################
## System 1 - Linear Sturm-Liouville ##
#######################################

# Create a library of forcing functions to use
i_set = [0.1,1.1]
j_set = [2,11]
k_set = [0.5,1.6]
s1_forcings = get_forcings(i_set=i_set, j_set=j_set, k_set=k_set)
s1_forcings = [f(x) for f in s1_forcings]
```

```python
##########################################
## System 2 - Nonlinear Sturm-Liouville ##
##########################################

# Create a library of forcing functions to use
i_set = [0.1,1.1]
j_set = [2,11]
k_set = [0.5,1.6]
s2_forcings = get_forcings(i_set=i_set, j_set=j_set, k_set=k_set)
s2_forcings = [f(x) for f in s2_forcings]
```

```python
###############################
## System 3 - Linear Poisson ##
###############################

# Get forcings
i_set = [0.2,0.7]
j_set = [1,10]
k_set = [0.7,1.3]
s3_forcings = get_forcings(i_set=i_set, j_set=j_set, k_set=k_set)
s3_forcings = [f(x) for f in s3_forcings]
```

```python
###############################
## System 4 - Bernoulli Beam ##
###############################

# Get forcings
i_set = [0.01,0.11]
j_set = [0.2,1.1]
k_set = [0.01,0.11]
s4_forcings = get_forcings(i_set=i_set, j_set=j_set, k_set=k_set)
s4_forcings = [f(x) for f in s4_forcings]
```

```python
fig = plt.figure(figsize=(10,5))
axs = fig.subplots(2, 2, sharex=True)

axs = axs.flatten()
forcings = [s1_forcings, s2_forcings, s3_forcings, s4_forcings]

for i, (ax, fs) in enumerate(zip(axs, forcings)):
    for f in fs:
        ax.plot(x, f)
        ax.set_title("Forcings: System {}".format(i+1))
        ax.set_ylabel('$f_j(x)$')

axs[3].set_xlabel('$x$')
axs[2].set_xlabel('$x$')
plt.tight_layout()
fig.savefig('SI-Forcings.png', dpi=1200)

plt.show()
```
